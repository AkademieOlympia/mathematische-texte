# Morph-Sandbox-Demo

Gleiche Logik wie `Morph.py`: Code in einer Sandbox ausführen und `stdout` auslesen.

**SageMath:** Die nächste Zelle importiert den Sage-Standard (`from sage.all import *`). Dafür brauchst du eine Umgebung mit installiertem SageMath oder wählst im Notebook den **SageMath 10.x**-Kernel (z. B. CoCalc / Sage-Distribution).

```python
# https://www.sagemath.org/ — üblicher Einstieg in Sage-Skripten und -Notebooks
from sage.all import *
```

```python
import asyncio
import sys
from pathlib import Path

# Modul morph_sandbox.py aus diesem Ordner laden (Notebook-CWD = Ordner des .ipynb)
_here = Path.cwd().resolve()
if str(_here) not in sys.path:
    sys.path.insert(0, str(_here))

from morph_sandbox import MorphSandbox


async def main():
    async with await MorphSandbox.create() as sandbox:
        await sandbox.execute_code("x = 42")
        result = await sandbox.execute_code("print(f'the answer is {x}')")
        print(result["output"])


await main()
```
