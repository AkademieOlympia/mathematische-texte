# Abstieg — Vierlinge und Lagrange (vier Quadrate)

Darstellung von $N = p_1 p_2 p_3 p_4$ als Summe von vier Quadraten (SymPy `sum_of_squares`) und Varianz der Beträge als grobes Streuungsmaß.

**Kernel:** normales Python 3 (SymPy muss installiert sein: `pip install sympy`).

```python
import math
from statistics import pvariance

from sympy.solvers.diophantine.diophantine import sum_of_squares


def analyze_vierling_lagrange(p_list):
    N = math.prod(p_list)
    # n = a^2 + b^2 + c^2 + d^2 (eine Lösung)
    sol = next(sum_of_squares(N, 4))
    variance = float(pvariance([abs(int(x)) for x in sol]))
    return sol, variance


# Beispiel: dichter vs. größerer Vierling (illustrativ)
v1 = [5, 7, 11, 13]
v2 = [101, 103, 107, 109]
```

```python
s1, var1 = analyze_vierling_lagrange(v1)
s2, var2 = analyze_vierling_lagrange(v2)
print("v1", v1, "N =", math.prod(v1), "->", s1, "Var(|.|) =", var1)
print("v2", v2, "N =", math.prod(v2), "->", s2, "Var(|.|) =", var2)
```
