```sage
# SageMath 10.5
from sage.all import *

# Definiere die Quaternionen-Algebra
Q.<i,j,k> = QuaternionAlgebra(QQ, -1, -1)

def hurwitz_norm(q):
    # Verwende reduced_norm() für die korrekte Norm-Berechnung
    n = q.reduced_norm()
    # Konvertiere zu Integer falls möglich
    if n in QQ:
        n_int = ZZ(n)
        if n_int == n:
            return n_int
    return n
```

```sage
# Erzeuge Hurwitz-Quaternionen kleiner Norm
# Hurwitz-Integers: entweder alle Koeffizienten ganz oder alle halb (ungerade/2)
primes = []
test_count = 0

# Optimierter Bereich - klein genug um schnell zu laufen, groß genug für viele Primzahlen
max_coord = 10  # Reduziert für bessere Performance

print("Berechne Hurwitz-Quaternionen...")
print(f"Suchbereich: -{max_coord} bis {max_coord} für ganzzahlige Koeffizienten")

# Fall 1: Ganzzahlige Hurwitz-Integers
for a in range(-max_coord, max_coord+1):
    for b in range(-max_coord, max_coord+1):
        for c in range(-max_coord, max_coord+1):
            for d in range(-max_coord, max_coord+1):
                q = a + b*i + c*j + d*k
                n = hurwitz_norm(q)
                test_count += 1
                if n in ZZ and n > 1 and is_prime(n):
                    primes.append((q, n))

print(f"Fall 1 abgeschlossen: {test_count} Quaternionen getestet")

# Fall 2: Halbe Hurwitz-Integers (alle Koeffizienten sind ungerade/2)
half_count = 0
for a in range(-(2*max_coord+1), 2*max_coord+2, 2):  # nur ungerade Zahlen
    for b in range(-(2*max_coord+1), 2*max_coord+2, 2):
        for c in range(-(2*max_coord+1), 2*max_coord+2, 2):
            for d in range(-(2*max_coord+1), 2*max_coord+2, 2):
                q = (a + b*i + c*j + d*k) / 2
                n = hurwitz_norm(q)
                half_count += 1
                if n in ZZ and n > 1 and is_prime(n):
                    primes.append((q, n))

test_count += half_count
print(f"Fall 2 abgeschlossen: {half_count} Quaternionen getestet")
print(f"Gesamt getestet: {test_count} Quaternionen")

# Extrahiere eindeutige Primzahlen (Normen) - verwende ein Dictionary um ein Beispiel-Quaternion zu behalten
prime_dict = {}
for q, n in primes:
    if n not in prime_dict:
        prime_dict[n] = q

# Sortiere nach Norm
unique_prime_norms = sorted(prime_dict.keys())

print(f"\n{'='*60}")
print(f"Hurwitz-Quaternionen mit primen Normen gefunden: {len(primes)}")
print(f"Eindeutige Hurwitz-Primzahlen (Normen) gefunden: {len(unique_prime_norms)}")
if unique_prime_norms:
    print(f"Erste 20 eindeutige Hurwitz-Primzahlen: {unique_prime_norms[:20]}")
    if len(unique_prime_norms) >= 100:
        print(f"✓ Genug Primzahlen gefunden für die ersten 100!")
    else:
        print(f"Hinweis: {len(unique_prime_norms)} eindeutige Primzahlen gefunden")
else:
    print("Keine Hurwitz-Primzahlen gefunden!")
```

```sage
# Grobe logarithmische Summe (analog Δ_defect)
log_sum = sum(1/log(p) for p in [p[1] for p in primes[:50]] if p > 2)
print("Approximative log-Dichte (erste 50):", float(log_sum))
```

```sage
# Ausgabe der ersten 100 eindeutigen Hurwitz-Primzahlen mit Koordinaten
# Verwende die bereits in Zelle 1 berechneten Variablen
if 'unique_prime_norms' not in globals() or 'prime_dict' not in globals():
    prime_dict = {}
    for q, n in primes:
        if n not in prime_dict:
            prime_dict[n] = q
    unique_prime_norms = sorted(prime_dict.keys())

def get_coords(q):
    """Extrahiert die Koordinaten (a, b, c, d) einer Quaternion"""
    try:
        coords = q.coefficient_tuple()
        if len(coords) >= 4:
            return coords[0], coords[1], coords[2], coords[3]
        elif len(coords) == 3:
            return coords[0], coords[1], coords[2], 0
        elif len(coords) == 2:
            return coords[0], coords[1], 0, 0
        else:
            return coords[0], 0, 0, 0
    except:
        return None, None, None, None

print("=" * 80)
print("ERSTE 100 EINDEUTIGE HURWITZ-PRIMZAHLEN MIT KOORDINATEN")
print("=" * 80)
print(f"\nGesamtanzahl eindeutiger Primzahlen: {len(unique_prime_norms)}")

# Kompakte Liste der Primzahlen
first_n = min(100, len(unique_prime_norms))
print(f"\nKompakte Liste der ersten {first_n} Primzahlen:")
print("-" * 80)
for i in range(0, first_n, 10):
    row = unique_prime_norms[i:i+10]
    print(" ".join(f"{p:4d}" for p in row))

# Detaillierte Liste mit Koordinaten
print(f"\n{'='*80}")
print(f"Detaillierte Liste mit Koordinaten (a, b, c, d) und Quaternionen:")
print("-" * 80)
print(f"{'Nr':<4} {'Primzahl':<10} {'Koordinaten (a,b,c,d)':<35} {'Quaternion':<30}")
print("-" * 80)

for idx, prime_norm in enumerate(unique_prime_norms[:first_n], 1):
    example_q = prime_dict[prime_norm]
    a, b, c, d = get_coords(example_q)
    if a is not None:
        coords_str = f"({a}, {b}, {c}, {d})"
    else:
        coords_str = "N/A"
    print(f"{idx:3d}. {prime_norm:8d} | {coords_str:<35} | {str(example_q):<30}")

print("\n" + "=" * 80)
print(f"Vollständige Liste der ersten {first_n} eindeutigen Primzahlen:")
print(unique_prime_norms[:first_n])
print("=" * 80)
```
