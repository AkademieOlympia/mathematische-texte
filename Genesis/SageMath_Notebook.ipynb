# SageMath Interaktives Notebook

Dieses Notebook verwendet den SageMath-Kernel für symbolische und numerische Mathematik.

## Grundlagen: Variablen und symbolische Ausdrücke

```sage
x = var('x')
expr = (x^2 - 1) / (x - 1)
show(expr)
show(expr.simplify_full())
```

## Gleichungen lösen

```sage
solve(x^2 - 3*x + 2 == 0, x)
```

## Analysis: Ableiten und Integrieren

```sage
f = sin(x) * exp(-x)
show(f.diff(x))
show(f.integral(x))
```

## Plotten

```sage
plot(sin(x), (x, -2*pi, 2*pi), aspect_ratio=1)
```

## Zahlentheorie

```sage
n = 123456789
factor(n)
is_prime(97)
```
