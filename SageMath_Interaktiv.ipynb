## 1. Primzahl-Analyse

Analysiere Primzahlen, Primfaktorzerlegungen und zahlentheoretische Eigenschaften.

```sage
# Primzahl-Analyse Widgets
prime_number_input = widgets.BoundedIntText(
    value=100,
    min=1,
    max=10**10,
    description='Zahl:',
    style={'description_width': 'initial'}
)

prime_analysis_output = widgets.Output()

def analyze_prime(n):
    """Analysiert eine Zahl auf Primzahl-Eigenschaften"""
    with prime_analysis_output:
        clear_output(wait=True)
        print(f"Analyse der Zahl: {n}\n")
        print("=" * 50)
        
        # Primzahl-Test
        is_prime_result = is_prime(n)
        print(f"Primzahl: {'✓ Ja' if is_prime_result else '✗ Nein'}")
        
        # Primfaktorzerlegung
        if n > 1:
            factors = factor(n)
            print(f"\nPrimfaktorzerlegung: {factors}")
            print(f"Anzahl der Primfaktoren: {len(factors)}")
        
        # Teiler
        divisors_list = divisors(n)
        print(f"\nTeiler ({len(divisors_list)}): {divisors_list}")
        
        # Eulersche Phi-Funktion
        if n > 0:
            phi_n = euler_phi(n)
            print(f"\nEulersche φ-Funktion: φ({n}) = {phi_n}")
        
        # Möbius-Funktion
        if n > 0:
            mu_n = moebius(n)
            print(f"Möbius-Funktion: μ({n}) = {mu_n}")
        
        # Quadratfreiheit
        is_sqfree = is_squarefree(n)
        print(f"Quadratfrei: {'✓ Ja' if is_sqfree else '✗ Nein'}")

def on_prime_analyze(b):
    n = prime_number_input.value
    analyze_prime(n)

prime_analyze_button = widgets.Button(
    description='Analysieren',
    button_style='primary',
    icon='search'
)
prime_analyze_button.on_click(on_prime_analyze)

prime_controls = widgets.VBox([
    widgets.HBox([prime_number_input, prime_analyze_button]),
    prime_analysis_output
])

display(prime_controls)
```

## 2. Primzahlen-Generator

Generiere Primzahlen in einem bestimmten Bereich.

```sage
# Primzahlen-Generator Widgets
prime_start_input = widgets.BoundedIntText(
    value=1,
    min=1,
    max=10**8,
    description='Von:',
    style={'description_width': 'initial'}
)

prime_end_input = widgets.BoundedIntText(
    value=100,
    min=1,
    max=10**8,
    description='Bis:',
    style={'description_width': 'initial'}
)

prime_count_input = widgets.BoundedIntText(
    value=20,
    min=1,
    max=1000,
    description='Anzahl:',
    style={'description_width': 'initial'}
)

prime_gen_output = widgets.Output()

def generate_primes(start, end, count):
    """Generiert Primzahlen"""
    with prime_gen_output:
        clear_output(wait=True)
        
        if count > 0:
            # N-te Primzahlen
            print(f"Die ersten {count} Primzahlen:\n")
            primes = [nth_prime(i) for i in range(1, count + 1)]
            print(primes)
            print(f"\nAnzahl: {len(primes)}")
        else:
            # Primzahlen im Bereich
            print(f"Primzahlen zwischen {start} und {end}:\n")
            primes = prime_range(start, end + 1)
            print(primes)
            print(f"\nAnzahl: {len(primes)}")
        
        if len(primes) > 0:
            print(f"\nKleinste: {primes[0]}")
            print(f"Größte: {primes[-1]}")

def on_prime_generate(b):
    start = prime_start_input.value
    end = prime_end_input.value
    count = prime_count_input.value
    generate_primes(start, end, count)

prime_gen_button = widgets.Button(
    description='Primzahlen generieren',
    button_style='success',
    icon='calculator'
)
prime_gen_button.on_click(on_prime_generate)

prime_gen_controls = widgets.VBox([
    widgets.HBox([prime_start_input, prime_end_input, prime_count_input]),
    prime_gen_button,
    prime_gen_output
])

display(prime_gen_controls)
```

## 3. Funktionen-Visualisierung

Visualisiere mathematische Funktionen interaktiv.

```sage
# Funktionen-Visualisierung Widgets
function_input = widgets.Text(
    value='sin(x)',
    description='Funktion:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)

x_min_input = widgets.FloatText(
    value=-10.0,
    description='x_min:',
    style={'description_width': 'initial'}
)

x_max_input = widgets.FloatText(
    value=10.0,
    description='x_max:',
    style={'description_width': 'initial'}
)

plot_output = widgets.Output()

def plot_function(func_str, x_min, x_max):
    """Plottet eine Funktion"""
    with plot_output:
        clear_output(wait=True)
        if not HAS_MATPLOTLIB:
            print("⚠️ Warnung: matplotlib ist nicht verfügbar!")
            print("Bitte installieren Sie matplotlib mit: pip install matplotlib")
            print("\nFunktion kann nicht visualisiert werden.")
            return
        
        try:
            x = var('x')
            func = sage_eval(func_str, locals={'x': x, 'sin': sin, 'cos': cos, 
                                                'tan': tan, 'exp': exp, 'log': log,
                                                'sqrt': sqrt, 'abs': abs})
            p = plot(func, (x, x_min, x_max), aspect_ratio='automatic')
            p.show()
            print(f"Funktion: {func_str}")
            print(f"Bereich: [{x_min}, {x_max}]")
        except Exception as e:
            print(f"Fehler beim Plotten: {e}")
            print("Verfügbare Funktionen: sin, cos, tan, exp, log, sqrt, abs")

def on_plot_function(b):
    func_str = function_input.value
    x_min = x_min_input.value
    x_max = x_max_input.value
    plot_function(func_str, x_min, x_max)

plot_button = widgets.Button(
    description='Plotten',
    button_style='info',
    icon='chart-line'
)
plot_button.on_click(on_plot_function)

# Warnung anzeigen, wenn matplotlib nicht verfügbar ist
warning_text = ""
if not HAS_MATPLOTLIB:
    warning_text = widgets.HTML(
        value='<div style="background-color: #fff3cd; padding: 10px; border: 1px solid #ffc107; border-radius: 5px; margin-bottom: 10px;">'
              '<strong>⚠️ Hinweis:</strong> matplotlib ist nicht verfügbar. Bitte installieren Sie es mit: <code>pip install matplotlib</code>'
              '</div>'
    )

plot_controls_list = [widgets.HBox([function_input, x_min_input, x_max_input]), plot_button, plot_output]
if warning_text:
    plot_controls_list.insert(0, warning_text)

plot_controls = widgets.VBox(plot_controls_list)

display(plot_controls)
```

## 4. Matrix-Operationen

Führe Matrix-Operationen durch: Determinante, Inverse, Eigenwerte, etc.

```sage
# Matrix-Operationen Widgets
matrix_size_input = widgets.BoundedIntText(
    value=3,
    min=1,
    max=10,
    description='Matrix-Größe:',
    style={'description_width': 'initial'}
)

matrix_output = widgets.Output()

def create_matrix_interface(size):
    """Erstellt eine interaktive Matrix-Eingabe"""
    with matrix_output:
        clear_output(wait=True)
        
        # Erstelle Eingabefelder für Matrix
        matrix_inputs = []
        for i in range(size):
            row = []
            for j in range(size):
                widget = widgets.FloatText(
                    value=0.0,
                    layout=widgets.Layout(width='80px'),
                    description=f'a{i+1}{j+1}:'
                )
                row.append(widget)
            matrix_inputs.append(row)
        
        # Matrix-Container
        matrix_box = widgets.VBox([
            widgets.HBox(row) for row in matrix_inputs
        ])
        
        # Ergebnis-Ausgabe
        result_output = widgets.Output()
        
        def compute_matrix_operations(b):
            """Berechnet Matrix-Operationen"""
            with result_output:
                clear_output(wait=True)
                try:
                    # Matrix aus Eingaben erstellen
                    matrix_data = []
                    for row in matrix_inputs:
                        matrix_data.append([w.value for w in row])
                    
                    M = matrix(QQ, matrix_data)
                    print("Matrix M:")
                    print(M)
                    print("\n" + "="*50)
                    
                    # Determinante
                    det = M.determinant()
                    print(f"\nDeterminante: det(M) = {det}")
                    
                    # Rang
                    rank = M.rank()
                    print(f"Rang: rank(M) = {rank}")
                    
                    # Inverse (falls invertierbar)
                    if det != 0:
                        M_inv = M.inverse()
                        print(f"\nInverse Matrix M⁻¹:")
                        print(M_inv)
                    else:
                        print("\nMatrix ist nicht invertierbar (Determinante = 0)")
                    
                    # Eigenwerte
                    try:
                        eigenvals = M.eigenvalues()
                        print(f"\nEigenwerte: {eigenvals}")
                    except:
                        print("\nEigenwerte konnten nicht berechnet werden")
                    
                    # Spur
                    trace = M.trace()
                    print(f"\nSpur: tr(M) = {trace}")
                    
                except Exception as e:
                    print(f"Fehler: {e}")
        
        compute_button = widgets.Button(
            description='Berechnen',
            button_style='warning',
            icon='calculator'
        )
        compute_button.on_click(compute_matrix_operations)
        
        display(widgets.VBox([
            widgets.Label("Matrix-Eingabe:"),
            matrix_box,
            compute_button,
            result_output
        ]))

def on_matrix_size_change(change):
    size = matrix_size_input.value
    create_matrix_interface(size)

matrix_size_input.observe(on_matrix_size_change, names='value')

# Initial anzeigen
create_matrix_interface(3)

matrix_controls = widgets.VBox([
    matrix_size_input,
    matrix_output
])

display(matrix_controls)
```

## 5. Zahlentheoretische Funktionen

Berechne verschiedene zahlentheoretische Funktionen.

```sage
# Zahlentheoretische Funktionen Widgets
nt_number_input = widgets.BoundedIntText(
    value=100,
    min=1,
    max=10**10,
    description='Zahl n:',
    style={'description_width': 'initial'}
)

nt_output = widgets.Output()

def compute_number_theory(n):
    """Berechnet zahlentheoretische Funktionen"""
    with nt_output:
        clear_output(wait=True)
        print(f"Zahlentheoretische Funktionen für n = {n}\n")
        print("=" * 60)
        
        # Teileranzahl
        sigma_0 = len(divisors(n))
        print(f"\nTeileranzahl: d(n) = σ₀(n) = {sigma_0}")
        
        # Teilersumme
        sigma_1 = sum(divisors(n))
        print(f"Teilersumme: σ(n) = σ₁(n) = {sigma_1}")
        
        # Eulersche Phi-Funktion
        phi_n = euler_phi(n)
        print(f"Eulersche φ-Funktion: φ(n) = {phi_n}")
        
        # Möbius-Funktion
        mu_n = moebius(n)
        print(f"Möbius-Funktion: μ(n) = {mu_n}")
        
        # Primfaktorzerlegung
        factors = factor(n)
        print(f"\nPrimfaktorzerlegung: {factors}")
        
        # Quadratfreiheit
        is_sqfree = is_squarefree(n)
        print(f"Quadratfrei: {'Ja' if is_sqfree else 'Nein'}")
        
        # Perfekte Zahl?
        if sigma_1 == 2*n:
            print(f"Perfekte Zahl: Ja (σ(n) = 2n)")
        elif sigma_1 > 2*n:
            print(f"Abundante Zahl: Ja (σ(n) > 2n)")
        else:
            print(f"Defiziente Zahl: Ja (σ(n) < 2n)")
        
        # Summe von Quadraten
        try:
            sum_of_squares = n.is_sum_of_two_squares()
            print(f"\nSumme von 2 Quadraten: {'Ja' if sum_of_squares else 'Nein'}")
        except:
            pass
        
        try:
            sum_of_four_squares = n.is_sum_of_four_squares()
            print(f"Summe von 4 Quadraten: {'Ja' if sum_of_four_squares else 'Nein'}")
        except:
            pass

def on_nt_compute(b):
    n = nt_number_input.value
    compute_number_theory(n)

nt_button = widgets.Button(
    description='Berechnen',
    button_style='primary',
    icon='calculator'
)
nt_button.on_click(on_nt_compute)

nt_controls = widgets.VBox([
    widgets.HBox([nt_number_input, nt_button]),
    nt_output
])

display(nt_controls)
```

## 6. Algebraische Berechnungen

Führe algebraische Operationen durch: Polynome, Gleichungen lösen, etc.

```sage
# Algebraische Berechnungen Widgets
poly_input = widgets.Text(
    value='x^2 - 5*x + 6',
    description='Polynom:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

algebra_output = widgets.Output()

def compute_algebra(poly_str):
    """Berechnet algebraische Eigenschaften"""
    with algebra_output:
        clear_output(wait=True)
        try:
            x = var('x')
            poly = sage_eval(poly_str, locals={'x': x})
            
            print(f"Polynom: {poly}\n")
            print("=" * 60)
            
            # Grad
            if hasattr(poly, 'degree'):
                deg = poly.degree()
                print(f"Grad: {deg}")
            
            # Nullstellen
            try:
                roots = poly.roots()
                print(f"\nNullstellen: {roots}")
            except:
                # Versuche symbolisch zu lösen
                try:
                    solutions = solve(poly == 0, x)
                    print(f"\nLösungen von {poly} = 0:")
                    for sol in solutions:
                        print(f"  {sol}")
                except:
                    print("\nNullstellen konnten nicht berechnet werden")
            
            # Faktorisierung
            try:
                factored = poly.factor()
                print(f"\nFaktorisierung: {factored}")
            except:
                print("\nFaktorisierung nicht möglich")
            
            # Ableitung
            try:
                derivative = poly.diff(x)
                print(f"\nAbleitung: {derivative}")
            except:
                pass
            
            # Stammfunktion
            try:
                integral = poly.integrate(x)
                print(f"\nStammfunktion: {integral}")
            except:
                pass
            
        except Exception as e:
            print(f"Fehler: {e}")
            print("\nBeispiel-Eingaben:")
            print("  x^2 - 5*x + 6")
            print("  x^3 + 2*x^2 - x - 2")
            print("  sin(x)")

def on_algebra_compute(b):
    poly_str = poly_input.value
    compute_algebra(poly_str)

algebra_button = widgets.Button(
    description='Berechnen',
    button_style='info',
    icon='calculator'
)
algebra_button.on_click(on_algebra_compute)

algebra_controls = widgets.VBox([
    widgets.HBox([poly_input, algebra_button]),
    algebra_output
])

display(algebra_controls)
```

## 8. Primzahl-Familien Visualisierung (MW-Winkel)

Visualisierung von Primzahl-Familien A, B, C, E mit geometrischer Abbildung und MW-Winkeltest.

```sage
# SageMath-kompatible Version des Primzahl-Familien Visualisierungs-Codes

# Imports - SageMath-kompatibel
from sage.all import *
import numpy as np

# Versuche matplotlib zu importieren
try:
    import matplotlib.pyplot as plt
    HAS_MATPLOTLIB = True
except ImportError:
    HAS_MATPLOTLIB = False
    print("Hinweis: matplotlib nicht verfügbar")

# ---------- Parameter ----------
N = 10**6
M = 10**6          # Referenzskala für Winkel
eps = 0.04         # Stärke der Glattheits-Radialverschiebung
THETA0 = 0.45      # MW-Winkel-Schwelle (rad)

# ---------- Primzahlen ----------
def primes_upto(n):
    """Generiert alle Primzahlen bis n mit SageMath prime_range"""
    # Verwende die effiziente SageMath-Funktion
    return list(prime_range(2, n + 1))

# ---------- Familien ----------
def fam(p):
    """Bestimmt die Familie einer Primzahl basierend auf p mod 12"""
    return {1:'A',5:'B',7:'C',11:'E'}.get(p%12,None)

# ---------- Glattheit ----------
# Cache für s23 und s235 (SageMath-Objekte sind hashable)
_s23_cache = {}
_s235_cache = {}

def s23(n):
    """Extrahiert die 2- und 3-Faktoren von n"""
    # Konvertiere zu Python int für Caching
    n_int = int(n)
    if n_int in _s23_cache:
        return _s23_cache[n_int]
    d = 1
    n_val = n_int
    while n_val % 2 == 0: 
        n_val //= 2
        d *= 2
    while n_val % 3 == 0: 
        n_val //= 3
        d *= 3
    _s23_cache[n_int] = d
    return d

def s235(n):
    """Extrahiert die 2-, 3- und 5-Faktoren von n"""
    n_int = int(n)
    if n_int in _s235_cache:
        return _s235_cache[n_int]
    d = s23(n)
    n_val = n_int
    while (n_val // d) % 5 == 0: 
        d *= 5
    _s235_cache[n_int] = d
    return d

def delta_smooth(n):
    """Berechnet die Glattheits-Differenz"""
    s23_val = s23(n)
    if s23_val > 0:
        return float(log(s235(n)/s23_val))
    else:
        return 0.0

# ---------- Geometrische Abbildung ----------
def theta_of(n):
    """Berechnet den Winkel für die geometrische Abbildung"""
    log_n = float(log(n))
    log_M = float(log(M))
    pi_val = float(pi)
    return 2 * pi_val * ((log_n % log_M) / log_M)

def r_of(n):
    """Berechnet den Radius mit Glattheits-Korrektur"""
    # Für Primzahlen: Verwende eine Kombination aus Glattheit und anderen Eigenschaften
    delta = delta_smooth(n)
    
    # Wenn delta = 0 (keine 2,3,5-Faktoren), verwende eine alternative Metrik
    # basierend auf n mod kleinen Primzahlen für Variation
    if delta == 0:
        # Verwende eine Variation basierend auf n mod 7, 11, 13 für visuelle Trennung
        # Dies gibt eine leichte Variation im Radius, auch wenn delta = 0
        variation = (n % 7 + n % 11 + n % 13) / 30.0  # Normalisiert zu [0, ~1]
        r = 0.75 + 0.20 * variation  # Radius zwischen 0.75 und 0.95
    else:
        r = max(0.75, 1 - eps*delta)
    
    return r

# ---------- Dreiecke (A,B,C) ----------
def triangle_vertices(phi):
    """Erstellt die Eckpunkte eines gleichseitigen Dreiecks"""
    pi_val = float(pi)
    return np.array([
        [float(cos(phi)), float(sin(phi))],
        [float(cos(phi + 2*pi_val/3)), float(sin(phi + 2*pi_val/3))],
        [float(cos(phi + 4*pi_val/3)), float(sin(phi + 4*pi_val/3))]
    ])

pi_val = float(pi)
TA = triangle_vertices(0.0)
TB = triangle_vertices(2*pi_val/3)
TC = triangle_vertices(4*pi_val/3)

# ---------- MW-Winkeltest ----------
def mw_angles(a,b,c):
    """Testet ob drei Primzahlen MW-kohärent sind"""
    ta, tb, tc = float(theta_of(a)), float(theta_of(b)), float(theta_of(c))
    pi_val = float(pi)
    two_pi = 2 * pi_val
    d = lambda x,y: abs(((x - y + pi_val) % two_pi) - pi_val)
    return max(d(ta,tb), d(tb,tc), d(tc,ta)) <= THETA0

# ---------- Hauptfunktion ----------
def plot_prime_families():
    """Hauptfunktion zur Visualisierung der Primzahl-Familien"""
    if not HAS_MATPLOTLIB:
        print("Fehler: matplotlib ist nicht verfügbar")
        return
    
    print("Berechne Primzahlen...")
    pr = primes_upto(N)
    print(f"Gefunden: {len(pr)} Primzahlen")
    
    print("Sortiere in Familien...")
    A = [p for p in pr if fam(p)=='A']
    B = [p for p in pr if fam(p)=='B']
    C = [p for p in pr if fam(p)=='C']
    E = [p for p in pr if fam(p)=='E']
    
    print(f"Familien: A={len(A)}, B={len(B)}, C={len(C)}, E={len(E)}")
    
    # ---------- Plot ----------
    fig,ax = plt.subplots(figsize=(8,8))
    ax.set_aspect('equal')
    ax.axis('off')
    
    # Kreis (E-Immanenz)
    t = np.linspace(0,2*pi,800)
    ax.plot(np.cos(t), np.sin(t), lw=1, color='black')
    
    # Dreiecke
    for T,col,lbl in [(TA,'tab:blue','A'),(TB,'tab:orange','B'),(TC,'tab:green','C')]:
        P = np.vstack([T,T[0]])
        ax.plot(P[:,0],P[:,1],lw=1.2,color=col,label=lbl)
    
    # Punkte (ausgedünnt für Lesbarkeit - mehr Punkte sichtbar)
    def plot_family(P, color, family_name):
        step = max(1, len(P)//500)  # Mehr Punkte: 500 statt 300 (kleinerer Divisor = mehr Punkte)
        count = 0
        radii = []
        for p in P[::step]:
            th, r = theta_of(p), r_of(p)
            radii.append(r)
            ax.scatter(r*float(cos(th)), r*float(sin(th)), s=8, color=color, alpha=0.8)  # Größere Punkte
            count += 1
        if len(radii) > 0:
            print(f"  {family_name}: {count} Punkte geplottet, Radius-Bereich: [{min(radii):.4f}, {max(radii):.4f}]")
        else:
            print(f"  {family_name}: Keine Punkte geplottet!")
    
    # Diagnose: Prüfe Radius-Verteilung für jede Familie
    def diagnose_radii(P, family_name):
        """Diagnostiziert die Radius-Verteilung"""
        sample_size = min(100, len(P))
        radii = [r_of(p) for p in P[:sample_size]]
        on_circle = sum(1 for r in radii if abs(r - 1.0) < 0.001)
        print(f"  {family_name} Diagnose (erste {sample_size}): {on_circle}/{sample_size} Punkte auf Kreisrand (r≈1.0)")
    
    print("Diagnose der Radius-Verteilung...")
    diagnose_radii(A, 'A')
    diagnose_radii(B, 'B')
    diagnose_radii(C, 'C')
    
    print("Plotte Familien A, B, C...")
    plot_family(A,'tab:blue', 'A')
    plot_family(B,'tab:orange', 'B')
    plot_family(C,'tab:green', 'C')
    
    # E nur auf dem Kreis (auf dem Kreisrand selbst) - IMMER Radius = 1
    print("Plotte Familie E...")
    step_e = max(1, len(E)//500)  # Gleiche Ausdünnung wie A, B, C
    count_e = 0
    # Plotte E-Punkte mit Label nur für den ersten Punkt
    first_e = True
    for e in E[::step_e]:
        th = theta_of(e)
        # E MUSS immer auf dem Kreisrand sein (Radius = 1, NICHT r_of(e))
        if first_e:
            ax.scatter(float(cos(th)), float(sin(th)), s=10, color='red', alpha=0.8, label='E (auf Kreisrand)', marker='o', edgecolors='darkred', linewidths=0.5)
            first_e = False
        else:
            ax.scatter(float(cos(th)), float(sin(th)), s=10, color='red', alpha=0.8, marker='o', edgecolors='darkred', linewidths=0.5)
        count_e += 1
    print(f"  E: {count_e} Punkte geplottet (alle auf Radius = 1.0)")
    
    # Beispiel: ein MW-kohärentes Tripel + Präquadrat-Skizze (heuristisch)
    print("Suche nach MW-kohärenten Tripeln...")
    found = False
    for a in A[-80:]:
        for b in B[-80:]:
            for c in C[-80:]:
                if mw_angles(a,b,c):
                    pts=[]
                    for x in (a,b,c):
                        th, r = theta_of(x), r_of(x)
                        pts.append([r*float(cos(th)), r*float(sin(th))])
                    pts.append(pts[0])
                    P=np.array(pts)
                    ax.plot(P[:,0],P[:,1],color='purple',lw=1.2, label='MW-Tripel' if not found else '')
                    print(f"MW-kohärentes Tripel gefunden: ({a}, {b}, {c})")
                    found = True
                    break
            if found:
                break
        if found:
            break
    
    if not found:
        print("Kein MW-kohärentes Tripel in den letzten 80 Primzahlen jeder Familie gefunden.")
    
    ax.legend(loc='upper right')
    plt.show()

# Führe die Visualisierung aus
plot_prime_families()
```

```sage
from sage.all import *
import matplotlib.pyplot as plt

def get_prime_class(p):
    """Bestimmt die Primzahl-Klasse basierend auf p mod 12"""
    m = p % 12
    return {5:'A', 7:'B', 11:'C', 1:'D'}.get(m, '?')

def calculate_pi_quat(t_win, p_win, eps=0.01, beta=0.01):
    """Berechnet das Quaternionen-Potential basierend auf Primzahl-Klassen"""
    N = len(t_win)
    mid = N // 2
    target_class = get_prime_class(p_win[mid])
    
    # R-Matrix nach Hamilton-Regeln
    R = matrix(RR, N, N)
    for i in range(N):
        for j in range(i+1, N):
            if i == mid or j == mid: continue
            c1, c2 = get_prime_class(p_win[i]), get_prime_class(p_win[j])
            if {c1, c2} == {'B', 'C'} and target_class == 'A': R[i,j]=1; R[j,i]=-1
            if {c1, c2} == {'A', 'C'} and target_class == 'B': R[i,j]=1; R[j,i]=-1
            if {c1, c2} == {'A', 'B'} and target_class == 'C': R[i,j]=1; R[j,i]=-1
                
    A = diagonal_matrix(RR, t_win) + eps * R
    lambdas = A.singular_values()
    
    # Entropie S und Stabilität Psi
    b = var('b')
    Z = 2 * sum(cosh(b * L) for L in lambdas)
    S_val = float( (log(Z) - b * diff(log(Z), b))(b=beta) )
    psi = 1.0 / (1.0 + S_val)
    
    # Resonanz-Check für Klassen-Boost
    classes = [get_prime_class(p) for p in p_win]
    res_boost = 1.2 if ('B' in classes and 'C' in classes and target_class == 'A') else 1.0
    
    return psi * res_boost

# Beispiel-Berechnung mit Testdaten
print("=" * 60)
print("Quaternionen-Potential Berechnung")
print("=" * 60)

# Beispiel: Verwende die Daten aus der vorherigen Zelle
t_example = [102.3, 105.446623052, 107.168611184, 111.029535543, 114.5]
p_example = [127, 131, 137, 139, 149]

print(f"\nEingabedaten:")
print(f"  t-Werte: {t_example}")
print(f"  Primzahlen: {p_example}")

# Bestimme Klassen
classes = [get_prime_class(p) for p in p_example]
print(f"\nPrimzahl-Klassen: {classes}")

# Berechne Potential
potential = calculate_pi_quat(t_example, p_example, eps=0.01, beta=0.01)
print(f"\nBerechnetes Quaternionen-Potential: {potential:.6f}")

# Zusätzliche Analyse
print(f"\nZusätzliche Informationen:")
print(f"  Ziel-Klasse (Mitte): {get_prime_class(p_example[len(p_example)//2])}")
print(f"  Anzahl der Klassen: {len(set(classes))}")

# Test mit verschiedenen Parametern
print(f"\n" + "=" * 60)
print("Sensitivitätsanalyse:")
print("=" * 60)
for eps_val in [0.001, 0.01, 0.1]:
    pot = calculate_pi_quat(t_example, p_example, eps=eps_val, beta=0.01)
    print(f"  ε = {eps_val:6.3f} → Potential = {pot:.6f}")
```

```sage
from sage.all import *

# Stelle sicher, dass get_prime_class und calculate_entropy verfügbar sind
# (sollten aus vorherigen Zellen geladen sein)
def get_prime_class(p):
    """Bestimmt die Primzahl-Klasse basierend auf p mod 12"""
    m = p % 12
    return {5:'A', 7:'B', 11:'C', 1:'D'}.get(m, '?')

def calculate_entropy(t_win, p_win, eps=0.01, beta=0.01):
    """Berechnet Entropie basierend auf t- und p-Werten"""
    N = len(t_win)
    mid = N // 2
    target_class = get_prime_class(p_win[mid])
    
    # R-Matrix nach Hamilton-Regeln
    R = matrix(RR, N, N)
    for i in range(N):
        for j in range(i+1, N):
            if i == mid or j == mid: continue
            c1, c2 = get_prime_class(p_win[i]), get_prime_class(p_win[j])
            if {c1, c2} == {'B', 'C'} and target_class == 'A': R[i,j]=1; R[j,i]=-1
            if {c1, c2} == {'A', 'C'} and target_class == 'B': R[i,j]=1; R[j,i]=-1
            if {c1, c2} == {'A', 'B'} and target_class == 'C': R[i,j]=1; R[j,i]=-1
                
    A = diagonal_matrix(RR, t_win) + eps * R
    lambdas = A.singular_values()
    
    # Entropie S
    b = var('b')
    Z = 2 * sum(cosh(b * L) for L in lambdas)
    S_val = float( (log(Z) - b * diff(log(Z), b))(b=beta) )
    
    return {'entropy': S_val}

def prime_state_equation(k, all_t, all_p, beta=0.01):
    """Berechnet theoretischen und realen Druck für eine Primzahl-Position
    mit Walter- und Kepler-Restriktionen"""
    # Sicherstellen, dass k im gültigen Bereich liegt
    if k < 2 or k >= len(all_p) - 1:
        raise ValueError(f"Index k={k} außerhalb des gültigen Bereichs [2, {len(all_p)-2}]")
    
    # 1. Berechne S(k) - Entropie im Fenster um k
    t_win = all_t[k-2:k+3]
    p_win = all_p[k-2:k+3]
    res = calculate_entropy(t_win, p_win, beta=beta)
    S = res['entropy']
    
    # 2. Bestimme Q (Klassen-Resonanz)
    p_k = all_p[k]
    cls = [p % 12 for p in all_p[k-2:k+3]]
    # Q ist hoch, wenn A, B und C im Fenster präsent sind
    Q = 1.0 + 0.5 * len(set(cls) - {1}) # Bonus für Vielfalt der Klassen A, B, C
    
    # 3. Berechne den theoretischen "Druck" (Dichte-Potential)
    T = 1.0 / beta
    
    # 4. KEPLER-RESTRIKTION: Maximale Packungsdichte π/(3√2) ≈ 0.74048
    # Dies ist die obere Schranke für die lokale Primzahldichte
    kepler_limit = float(pi / (3 * sqrt(2)))  # ≈ 0.74048
    
    # 5. WALTER-RESTRIKTION: Entropie-basierte Gültigkeitsbedingung
    # Die Zustandsgleichung ist nur gültig, wenn die Entropie innerhalb
    # eines bestimmten Bereichs liegt (Stabilitätsbedingung)
    S_min = 2.0  # Minimale Entropie für Stabilität
    S_max = 2.5  # Maximale Entropie für Gültigkeit
    walter_valid = (S_min <= S <= S_max)
    
    # 6. Verbesserte Druck-Formel mit Kepler-Limit
    # P_theory wird durch das Kepler-Limit skaliert und begrenzt
    P_theory_raw = T * exp(-S / Q)
    
    # 7. KEPLER-RESTRIKTION: Anwendung des maximalen Packungsdichte-Limits
    # Das Kepler-Limit π/(3√2) ≈ 0.74048 ist die maximale Packungsdichte
    # Wir verwenden es als Skalierungsfaktor für den theoretischen Druck
    # P_max = T * kepler_limit (maximaler theoretischer Druck basierend auf Kepler-Limit)
    P_max_kepler = T * kepler_limit
    
    # Skaliere P_theory_raw mit dem Kepler-Limit-Faktor
    # Dies begrenzt den Druck auf realistische Werte
    P_theory_scaled = P_theory_raw * kepler_limit
    
    # 8. WALTER-RESTRIKTION: Entropie-basierte Korrektur
    # Wenn Entropie außerhalb des gültigen Bereichs, wird der Druck angepasst
    if not walter_valid:
        # Korrekturfaktor basierend auf Entropie-Abweichung
        if S < S_min:
            # Zu niedrige Entropie: Druck wird leicht erhöht (weniger Ordnung = höherer Druck)
            correction = 1.0 + 0.15 * (S_min - S)
        else:  # S > S_max
            # Zu hohe Entropie: Druck wird reduziert (mehr Unordnung = niedrigerer Druck)
            correction = 1.0 - 0.15 * (S - S_max)
        correction = max(0.2, min(1.5, correction))  # Begrenze Korrektur auf sinnvolle Werte
        P_theory = P_theory_scaled * correction
    else:
        # Wenn Walter-Restriktion erfüllt ist, verwende den skalierten Wert direkt
        P_theory = P_theory_scaled
    
    # 9. Realer lokaler Druck (Reziproke des Gaps)
    gap = all_p[k+1] - all_p[k]
    P_real = 1.0 / gap if gap > 0 else float('inf')
    
    return P_theory, P_real, S, Q, walter_valid, kepler_limit

# Beispiel-Berechnung
print("=" * 60)
print("Primzahl-Zustandsgleichung")
print("=" * 60)

# Erstelle Beispiel-Daten
primes_example = list(prime_range(100, 200))
t_example = [float(log(p)) * 10 for p in primes_example]

print(f"\nVerwende {len(primes_example)} Primzahlen im Bereich [{primes_example[0]}, {primes_example[-1]}]")

# Teste für verschiedene Indizes
test_indices = [10, 20, 30, 40] if len(primes_example) > 40 else list(range(2, min(10, len(primes_example)-1)))

print(f"\nBerechne Zustandsgleichung für Indizes: {test_indices}")
print("-" * 60)

print("\n" + "=" * 60)
print("VERGLEICH: Mit vs. Ohne Walter- und Kepler-Restriktionen")
print("=" * 60)

for k in test_indices:
    try:
        # Neue Berechnung mit Restriktionen
        P_theory_new, P_real, S, Q, walter_valid, kepler_limit = prime_state_equation(k, t_example, primes_example, beta=0.01)
        
        # Alte Berechnung ohne Restriktionen (zum Vergleich)
        T = 1.0 / 0.01
        P_theory_old = T * exp(-S / Q)
        
        gap = primes_example[k+1] - primes_example[k]
        ratio_old = P_theory_old / P_real if P_real > 0 else float('inf')
        ratio_new = P_theory_new / P_real if P_real > 0 else float('inf')
        
        print(f"\nk={k}: p[{k}]={primes_example[k]}, p[{k+1}]={primes_example[k+1]} (Gap={gap})")
        print(f"  Entropie S = {S:.6f}")
        print(f"  Klassen-Resonanz Q = {Q:.3f}")
        print(f"  Kepler-Limit (max. Packungsdichte) = {kepler_limit:.6f}")
        print(f"  Walter-Restriktion gültig: {'✓ Ja' if walter_valid else '✗ Nein'}")
        print(f"\n  OHNE Restriktionen:")
        print(f"    P_theory_old = {P_theory_old:.6f}")
        print(f"    Verhältnis P_theory_old/P_real = {ratio_old:.6f}")
        print(f"\n  MIT Walter- und Kepler-Restriktionen:")
        print(f"    P_theory_new = {P_theory_new:.6f}")
        print(f"    Verhältnis P_theory_new/P_real = {ratio_new:.6f}")
        print(f"    Realer Druck P_real = {P_real:.6f}")
        
        # Berechne Verbesserung
        improvement = abs(ratio_old - 1.0) - abs(ratio_new - 1.0)
        if improvement > 0:
            print(f"\n  ✓ Verbesserung: {improvement:.6f} (näher an 1.0)")
        else:
            print(f"\n  → Anpassung erforderlich")
            
    except (ValueError, IndexError) as e:
        print(f"\nk={k}: Fehler - {e}")

print("\n" + "=" * 60)
```

```sage
from sage.all import *
import numpy as np

# ---------- Morley-Walter 3-Kohärenz: Verbesserte Implementierung ----------
def fam_mw(p):
    """Primzahl-Familie basierend auf p mod 12"""
    r = p % 12
    return {1: 'E', 5: 'A', 7: 'B', 11: 'C'}.get(r, None)

def quat_target_sign_mw(c1, c2, target):
    """Quaternionische Orientierung mit Zielklasse (Morley-Walter)"""
    if c1 is None or c2 is None or target is None:
        return 0
    if 'E' in (c1, c2, target):
        return 0
    
    # Forward products: A*B=C, B*C=A, C*A=B
    forward = {('A','B'):'C', ('B','C'):'A', ('C','A'):'B'}
    
    if forward.get((c1, c2), None) == target:
        return +1
    if forward.get((c2, c1), None) == target:
        return -1
    return 0

class LocalPrimeDiracMW:
    """Lokaler Primzahl-Dirac-Operator mit Morley-Walter 3-Kohärenz"""
    def __init__(self, primes_list, center_index, window_m=25,
                 xi=1.0, idx_sigma=10.0, use_idx_weight=True):
        self.primes = primes_list
        self.N = len(self.primes)
        self.k = center_index
        self.m = window_m
        self.xi = float(xi)
        self.idx_sigma = float(idx_sigma)
        self.use_idx_weight = bool(use_idx_weight)
        
        # Fenster-Indizes
        lo = max(0, self.k - self.m)
        hi = min(self.N - 1, self.k + self.m)
        self.idxs = list(range(lo, hi + 1))
        self.n = len(self.idxs)
        self.p_center = self.primes[self.k]
        self.c_center = fam_mw(self.p_center)
    
    def idx_weight(self, i, j, k):
        """Gewichtung basierend auf Entfernung vom Zentrum"""
        if not self.use_idx_weight:
            return 1.0
        # Glatte "Kohärenz-Radius" um den zentralen Index k
        di = abs(i - k)
        dj = abs(j - k)
        return float(exp(-(di*di + dj*dj)/(2.0*self.idx_sigma*self.idx_sigma)))
    
    def build_A(self):
        """Konstruiere A = diag(log p) + K mit schiefsymmetrischem K"""
        ps = [self.primes[i] for i in self.idxs]
        cs = [fam_mw(p) for p in ps]
        diag = [float(log(p)) for p in ps]
        
        K = matrix(RR, self.n, self.n)
        
        for a in range(self.n):
            i = self.idxs[a]
            pi = ps[a]
            ci = cs[a]
            
            for b in range(a+1, self.n):
                j = self.idxs[b]
                pj = ps[b]
                cj = cs[b]
                
                sgn = quat_target_sign_mw(ci, cj, self.c_center)
                if sgn == 0:
                    continue
                
                # Mangoldt-ähnliche Interaktion
                interaction = float((log(pi) * log(pj)) / sqrt(pi * pj))
                
                w = self.idx_weight(i, j, self.k)
                val = self.xi * interaction * w * sgn
                
                K[a, b] = val
                K[b, a] = -val  # schiefsymmetrisch
        
        # A = diag(log p) + K
        A = diagonal_matrix(RR, diag) + K
        return A
    
    def build_dirac(self):
        """Konstruiere Dirac-Operator D = [[0, A], [A^T, 0]]"""
        A = self.build_A()
        n = A.nrows()
        Z = zero_matrix(RR, n, n)
        
        # Dirac-Operator als Block-Matrix
        D = block_matrix([[Z, A], [A.transpose(), Z]])
        return D
    
    def get_entropy_from_dirac(self, beta=0.01):
        """Berechne Entropie aus dem Dirac-Operator-Spektrum"""
        D = self.build_dirac()
        # Eigenwerte des Dirac-Operators
        eigvals = D.eigenvalues()
        eigvals = [float(abs(e)) for e in eigvals]  # Beträge
        
        # Entropie-Berechnung
        b = var('b')
        Z = 2 * sum(cosh(b * L) for L in eigvals)
        S_val = float((log(Z) - b * diff(log(Z), b))(b=beta))
        
        return {'entropy': S_val, 'eigenvalues': eigvals}

# ---------- Verbesserte Zustandsgleichung mit Morley-Walter ----------
def prime_state_equation_mw(k, all_t, all_p, beta=0.01, window_m=25, 
                            xi=1.0, idx_sigma=10.0):
    """Berechnet theoretischen und realen Druck mit Morley-Walter 3-Kohärenz"""
    if k < window_m or k >= len(all_p) - window_m:
        raise ValueError(f"Index k={k} außerhalb des gültigen Bereichs")
    
    # 1. Erstelle lokalen Dirac-Operator mit Morley-Walter
    model = LocalPrimeDiracMW(all_p, k, window_m=window_m, 
                              xi=xi, idx_sigma=idx_sigma, use_idx_weight=True)
    res = model.get_entropy_from_dirac(beta=beta)
    S = res['entropy']
    
    # 2. Bestimme Q (Klassen-Resonanz) im Fenster
    cls = [fam_mw(p) for p in all_p[k-window_m:k+window_m+1]]
    # Q ist hoch, wenn A, B und C im Fenster präsent sind
    unique_classes = set(cls) - {None, 'E'}
    Q = 1.0 + 0.5 * len(unique_classes)
    
    # 3. KEPLER-RESTRIKTION: Maximale Packungsdichte
    kepler_limit = float(pi / (3 * sqrt(2)))  # ≈ 0.74048
    
    # 4. WALTER-RESTRIKTION: Entropie-basierte Gültigkeitsbedingung
    S_min = 2.0
    S_max = 2.5
    walter_valid = (S_min <= S <= S_max)
    
    # 5. Verbesserte Druck-Formel
    T = 1.0 / beta
    P_theory_raw = T * exp(-S / Q)
    
    # Anwendung des Kepler-Limits
    P_theory_scaled = P_theory_raw * kepler_limit
    
    # Walter-Restriktion: Entropie-basierte Korrektur
    if not walter_valid:
        if S < S_min:
            correction = 1.0 + 0.15 * (S_min - S)
        else:  # S > S_max
            correction = 1.0 - 0.15 * (S - S_max)
        correction = max(0.2, min(1.5, correction))
        P_theory = P_theory_scaled * correction
    else:
        P_theory = P_theory_scaled
    
    # 6. Realer lokaler Druck
    gap = all_p[k+1] - all_p[k]
    P_real = 1.0 / gap if gap > 0 else float('inf')
    
    return P_theory, P_real, S, Q, walter_valid, kepler_limit

# ---------- Test der verbesserten Zustandsgleichung ----------
print("=" * 70)
print("VERBESSERTE PRIMZAHL-ZUSTANDSGLEICHUNG")
print("Mit Morley-Walter 3-Kohärenz und lokaler Dirac-Struktur")
print("=" * 70)

# Erstelle Beispiel-Daten
primes_example = list(prime_range(100, 500))
t_example = [float(log(p)) * 10 for p in primes_example]

print(f"\nVerwende {len(primes_example)} Primzahlen im Bereich [{primes_example[0]}, {primes_example[-1]}]")

# Teste für verschiedene Indizes (mit ausreichendem Abstand)
test_indices = [30, 50, 70, 90] if len(primes_example) > 100 else list(range(25, min(50, len(primes_example)-25), 10))

print(f"\nBerechne verbesserte Zustandsgleichung für Indizes: {test_indices}")
print("-" * 70)

for k in test_indices:
    try:
        # Neue Berechnung mit Morley-Walter
        P_theory_mw, P_real, S_mw, Q_mw, walter_valid, kepler_limit = \
            prime_state_equation_mw(k, t_example, primes_example, 
                                   beta=0.01, window_m=25, xi=1.0, idx_sigma=10.0)
        
        # Alte Berechnung ohne Morley-Walter (zum Vergleich)
        P_theory_old, P_real_old, S_old, Q_old, _, _ = \
            prime_state_equation(k, t_example, primes_example, beta=0.01)
        
        gap = primes_example[k+1] - primes_example[k]
        ratio_old = P_theory_old / P_real if P_real > 0 else float('inf')
        ratio_mw = P_theory_mw / P_real if P_real > 0 else float('inf')
        
        print(f"\nk={k}: p[{k}]={primes_example[k]}, p[{k+1}]={primes_example[k+1]} (Gap={gap})")
        print(f"  Entropie (alt): S = {S_old:.6f}")
        print(f"  Entropie (MW):  S = {S_mw:.6f}")
        print(f"  Klassen-Resonanz Q = {Q_mw:.3f}")
        print(f"  Kepler-Limit = {kepler_limit:.6f}")
        print(f"  Walter-Restriktion: {'✓ Ja' if walter_valid else '✗ Nein'}")
        
        print(f"\n  OHNE Morley-Walter:")
        print(f"    P_theory_old = {P_theory_old:.6f}")
        print(f"    Verhältnis = {ratio_old:.6f}")
        
        print(f"\n  MIT Morley-Walter 3-Kohärenz:")
        print(f"    P_theory_mw = {P_theory_mw:.6f}")
        print(f"    Verhältnis = {ratio_mw:.6f}")
        print(f"    P_real = {P_real:.6f}")
        
        # Berechne Verbesserung
        improvement = abs(ratio_old - 1.0) - abs(ratio_mw - 1.0)
        if improvement > 0:
            print(f"\n  ✓ Verbesserung: {improvement:.6f} (näher an 1.0)")
        else:
            print(f"\n  → Weitere Optimierung möglich")
            
    except (ValueError, IndexError) as e:
        print(f"\nk={k}: Fehler - {e}")

print("\n" + "=" * 70)
```

```sage
from sage.all import *
import numpy as np
import matplotlib.pyplot as plt

class GlobalPrimeHamiltonian:
    def __init__(self, limit=500, coupling_strength=0.5):
        # Wir nehmen alle Primzahlen bis 'limit'
        self.primes = list(primes(limit))
        self.N = len(self.primes)
        self.xi = coupling_strength
        
    def get_quat_sign(self, p1, p2):
        # Bestimme Klassen
        m1, m2 = p1 % 12, p2 % 12
        map_cls = {5:'A', 7:'B', 11:'C', 1:'D'}
        c1, c2 = map_cls.get(m1), map_cls.get(m2)
        
        # Quaternionische Regel (Beispielhaft für Wechselwirkung)
        # Wir definieren eine zyklische Orientierung A->B->C->A
        if c1 == 'A' and c2 == 'B': return 1
        if c1 == 'B' and c2 == 'A': return -1
        
        if c1 == 'B' and c2 == 'C': return 1
        if c1 == 'C' and c2 == 'B': return -1
        
        if c1 == 'C' and c2 == 'A': return 1
        if c1 == 'A' and c2 == 'C': return -1
        
        return 0 # Keine Wechselwirkung

    def build_matrix(self):
        H = np.zeros((self.N, self.N))
        
        for i in range(self.N):
            p_i = self.primes[i]
            H[i, i] = float(log(p_i))  # Diagonale: log(p)
            
            for j in range(i+1, self.N):
                p_j = self.primes[j]
                sign = self.get_quat_sign(p_i, p_j)
                if sign != 0:
                    H[i, j] = self.xi * sign
                    H[j, i] = -self.xi * sign  # Schiefsymmetrisch
        
        return H

# Konstruktion
builder = GlobalPrimeHamiltonian(limit=1000, coupling_strength=1.0)
H_prime = builder.build_matrix()

# Da wir ein schiefsymmetrisches R gebaut haben (ähnlich wie i*H),
# sind die Eigenwerte rein imaginär (oder +/- Paare bei Dirac).
# Wir betrachten die Beträge der Imaginärteile als "Energien".
eigenvalues = np.imag(np.linalg.eigvals(H_prime))
eigenvalues = np.sort(np.abs(eigenvalues))

# Analyse der Abstände (Level Spacing)
diffs = np.diff(eigenvalues)
# Normierung auf Mittelwert 1
diffs_norm = diffs / np.mean(diffs)

# Plot der Statistik
plt.figure(figsize=(10, 6))
plt.hist(diffs_norm, bins=30, density=True, alpha=0.7, color='purple', label='Eigenwert-Abstände (Primzahl-Hamiltonian)')

# Wigner-Surmise (GUE) zum Vergleich - Die Signatur der Riemann-Nullstellen
x = np.linspace(0, 4, 100)
p_wigner = (32/np.pi**2) * (x**2) * np.exp(-4*x**2/np.pi)
plt.plot(x, p_wigner, 'r--', label='Wigner-Surmise (GUE - Riemann-Signatur)')

plt.title('Emergenz der Riemann-Statistik aus reinem Primzahl-Input')
plt.xlabel('Normierter Abstand s')
plt.ylabel('P(s)')
plt.legend()
plt.savefig('global_operator_spectrum.png')
plt.show()
print("Spektral-Analyse gespeichert.")

# ---------- Hilfsfunktionen für prime_state_equation ----------
def get_prime_class(p):
    """Bestimmt die Primzahl-Klasse basierend auf p mod 12"""
    m = p % 12
    return {5:'A', 7:'B', 11:'C', 1:'D'}.get(m, '?')

def calculate_entropy(t_win, p_win, eps=0.01, beta=0.01):
    """Berechnet Entropie basierend auf t- und p-Werten"""
    N = len(t_win)
    mid = N // 2
    target_class = get_prime_class(p_win[mid])
    
    # R-Matrix nach Hamilton-Regeln
    R = matrix(RR, N, N)
    for i in range(N):
        for j in range(i+1, N):
            if i == mid or j == mid: continue
            c1, c2 = get_prime_class(p_win[i]), get_prime_class(p_win[j])
            if {c1, c2} == {'B', 'C'} and target_class == 'A': R[i,j]=1; R[j,i]=-1
            if {c1, c2} == {'A', 'C'} and target_class == 'B': R[i,j]=1; R[j,i]=-1
            if {c1, c2} == {'A', 'B'} and target_class == 'C': R[i,j]=1; R[j,i]=-1
                
    A = diagonal_matrix(RR, t_win) + eps * R
    lambdas = A.singular_values()
    
    # Entropie S
    b = var('b')
    Z = 2 * sum(cosh(b * L) for L in lambdas)
    S_val = float( (log(Z) - b * diff(log(Z), b))(b=beta) )
    
    return {'entropy': S_val}

def prime_state_equation(k, all_t, all_p, beta=0.01):
    """Berechnet theoretischen und realen Druck für eine Primzahl-Position
    mit Walter- und Kepler-Restriktionen"""
    # Sicherstellen, dass k im gültigen Bereich liegt (Fenster: k-2 bis k+3)
    if k < 2 or k >= len(all_p) - 3:
        raise ValueError(f"Index k={k} außerhalb des gültigen Bereichs [2, {len(all_p)-4}]")
    
    # 1. Berechne S(k) - Entropie im Fenster um k
    t_win = all_t[k-2:k+3]
    p_win = all_p[k-2:k+3]
    res = calculate_entropy(t_win, p_win, beta=beta)
    S = res['entropy']
    
    # 2. Bestimme Q (Klassen-Resonanz)
    p_k = all_p[k]
    cls = [p % 12 for p in all_p[k-2:k+3]]
    # Q ist hoch, wenn A, B und C im Fenster präsent sind
    Q = 1.0 + 0.5 * len(set(cls) - {1}) # Bonus für Vielfalt der Klassen A, B, C
    
    # 3. Berechne den theoretischen "Druck" (Dichte-Potential)
    T = 1.0 / beta
    
    # 4. KEPLER-RESTRIKTION: Maximale Packungsdichte π/(3√2) ≈ 0.74048
    kepler_limit = float(pi / (3 * sqrt(2)))  # ≈ 0.74048
    
    # 5. WALTER-RESTRIKTION: Entropie-basierte Gültigkeitsbedingung
    S_min = 2.0  # Minimale Entropie für Stabilität
    S_max = 2.5  # Maximale Entropie für Gültigkeit
    walter_valid = (S_min <= S <= S_max)
    
    # 6. Verbesserte Druck-Formel mit Kepler-Limit
    P_theory_raw = T * exp(-S / Q)
    
    # 7. KEPLER-RESTRIKTION: Anwendung des maximalen Packungsdichte-Limits
    P_theory_scaled = P_theory_raw * kepler_limit
    
    # 8. WALTER-RESTRIKTION: Entropie-basierte Korrektur
    if not walter_valid:
        if S < S_min:
            correction = 1.0 + 0.15 * (S_min - S)
        else:  # S > S_max
            correction = 1.0 - 0.15 * (S - S_max)
        correction = max(0.2, min(1.5, correction))
        P_theory = P_theory_scaled * correction
    else:
        P_theory = P_theory_scaled
    
    # 9. Realer lokaler Druck (Reziproke des Gaps)
    gap = all_p[k+1] - all_p[k]
    P_real = 1.0 / gap if gap > 0 else float('inf')
    
    return P_theory, P_real, S, Q, walter_valid, kepler_limit

# ---------- Morley-Walter 3-Kohärenz: Verbesserte Implementierung ----------
def fam_mw(p):
    """Primzahl-Familie basierend auf p mod 12"""
    r = p % 12
    return {1: 'E', 5: 'A', 7: 'B', 11: 'C'}.get(r, None)

def quat_target_sign_mw(c1, c2, target):
    """Quaternionische Orientierung mit Zielklasse (Morley-Walter)"""
    if c1 is None or c2 is None or target is None:
        return 0
    if 'E' in (c1, c2, target):
        return 0
    
    # Forward products: A*B=C, B*C=A, C*A=B
    forward = {('A','B'):'C', ('B','C'):'A', ('C','A'):'B'}
    
    if forward.get((c1, c2), None) == target:
        return +1
    if forward.get((c2, c1), None) == target:
        return -1
    return 0

class LocalPrimeDiracMW:
    """Lokaler Primzahl-Dirac-Operator mit Morley-Walter 3-Kohärenz"""
    def __init__(self, primes_list, center_index, window_m=25,
                 xi=1.0, idx_sigma=10.0, use_idx_weight=True):
        self.primes = primes_list
        self.N = len(self.primes)
        self.k = center_index
        self.m = window_m
        self.xi = float(xi)
        self.idx_sigma = float(idx_sigma)
        self.use_idx_weight = bool(use_idx_weight)
        
        # Fenster-Indizes
        lo = max(0, self.k - self.m)
        hi = min(self.N - 1, self.k + self.m)
        self.idxs = list(range(lo, hi + 1))
        self.n = len(self.idxs)
        self.p_center = self.primes[self.k]
        self.c_center = fam_mw(self.p_center)
    
    def idx_weight(self, i, j, k):
        """Gewichtung basierend auf Entfernung vom Zentrum"""
        if not self.use_idx_weight:
            return 1.0
        # Glatte "Kohärenz-Radius" um den zentralen Index k
        di = abs(i - k)
        dj = abs(j - k)
        return float(exp(-(di*di + dj*dj)/(2.0*self.idx_sigma*self.idx_sigma)))
    
    def build_A(self):
        """Konstruiere A = diag(log p) + K mit schiefsymmetrischem K"""
        ps = [self.primes[i] for i in self.idxs]
        cs = [fam_mw(p) for p in ps]
        diag = [float(log(p)) for p in ps]
        
        K = matrix(RR, self.n, self.n)
        
        for a in range(self.n):
            i = self.idxs[a]
            pi = ps[a]
            ci = cs[a]
            
            for b in range(a+1, self.n):
                j = self.idxs[b]
                pj = ps[b]
                cj = cs[b]
                
                sgn = quat_target_sign_mw(ci, cj, self.c_center)
                if sgn == 0:
                    continue
                
                # Mangoldt-ähnliche Interaktion
                interaction = float((log(pi) * log(pj)) / sqrt(pi * pj))
                
                w = self.idx_weight(i, j, self.k)
                val = self.xi * interaction * w * sgn
                
                K[a, b] = val
                K[b, a] = -val  # schiefsymmetrisch
        
        # A = diag(log p) + K
        A = diagonal_matrix(RR, diag) + K
        return A
    
    def build_dirac(self):
        """Konstruiere Dirac-Operator D = [[0, A], [A^T, 0]]"""
        A = self.build_A()
        n = A.nrows()
        Z = zero_matrix(RR, n, n)
        
        # Dirac-Operator als Block-Matrix
        D = block_matrix([[Z, A], [A.transpose(), Z]])
        return D
    
    def get_entropy_from_dirac(self, beta=0.01):
        """Berechne Entropie aus dem Dirac-Operator-Spektrum"""
        D = self.build_dirac()
        # Eigenwerte des Dirac-Operators
        eigvals = D.eigenvalues()
        eigvals = [float(abs(e)) for e in eigvals]  # Beträge
        
        # Entropie-Berechnung
        b = var('b')
        Z = 2 * sum(cosh(b * L) for L in eigvals)
        S_val = float((log(Z) - b * diff(log(Z), b))(b=beta))
        
        return {'entropy': S_val, 'eigenvalues': eigvals}

# ---------- Verbesserte Zustandsgleichung mit Morley-Walter ----------
def prime_state_equation_mw(k, all_t, all_p, beta=0.01, window_m=25, 
                            xi=1.0, idx_sigma=10.0):
    """Berechnet theoretischen und realen Druck mit Morley-Walter 3-Kohärenz"""
    if k < window_m or k >= len(all_p) - window_m:
        raise ValueError(f"Index k={k} außerhalb des gültigen Bereichs")
    
    # 1. Erstelle lokalen Dirac-Operator mit Morley-Walter
    model = LocalPrimeDiracMW(all_p, k, window_m=window_m, 
                              xi=xi, idx_sigma=idx_sigma, use_idx_weight=True)
    res = model.get_entropy_from_dirac(beta=beta)
    S = res['entropy']
    
    # 2. Bestimme Q (Klassen-Resonanz) im Fenster
    cls = [fam_mw(p) for p in all_p[k-window_m:k+window_m+1]]
    # Q ist hoch, wenn A, B und C im Fenster präsent sind
    unique_classes = set(cls) - {None, 'E'}
    Q = 1.0 + 0.5 * len(unique_classes)
    
    # 3. KEPLER-RESTRIKTION: Maximale Packungsdichte
    kepler_limit = float(pi / (3 * sqrt(2)))  # ≈ 0.74048
    
    # 4. WALTER-RESTRIKTION: Entropie-basierte Gültigkeitsbedingung
    S_min = 2.0
    S_max = 2.5
    walter_valid = (S_min <= S <= S_max)
    
    # 5. Verbesserte Druck-Formel
    T = 1.0 / beta
    P_theory_raw = T * exp(-S / Q)
    
    # Anwendung des Kepler-Limits
    P_theory_scaled = P_theory_raw * kepler_limit
    
    # Walter-Restriktion: Entropie-basierte Korrektur
    if not walter_valid:
        if S < S_min:
            correction = 1.0 + 0.15 * (S_min - S)
        else:  # S > S_max
            correction = 1.0 - 0.15 * (S - S_max)
        correction = max(0.2, min(1.5, correction))
        P_theory = P_theory_scaled * correction
    else:
        P_theory = P_theory_scaled
    
    # 6. Realer lokaler Druck
    gap = all_p[k+1] - all_p[k]
    P_real = 1.0 / gap if gap > 0 else float('inf')
    
    return P_theory, P_real, S, Q, walter_valid, kepler_limit

# ---------- Test der verbesserten Zustandsgleichung ----------
print("=" * 70)
print("VERBESSERTE PRIMZAHL-ZUSTANDSGLEICHUNG")
print("Mit Morley-Walter 3-Kohärenz und lokaler Dirac-Struktur")
print("=" * 70)

# Erstelle Beispiel-Daten
primes_example = list(prime_range(100, 500))
t_example = [float(log(p)) * 10 for p in primes_example]

print(f"\nVerwende {len(primes_example)} Primzahlen im Bereich [{primes_example[0]}, {primes_example[-1]}]")

# Teste für verschiedene Indizes (mit ausreichendem Abstand)
test_indices = [30, 50, 70, 90] if len(primes_example) > 100 else list(range(25, min(50, len(primes_example)-25), 10))

print(f"\nBerechne verbesserte Zustandsgleichung für Indizes: {test_indices}")
print("-" * 70)

for k in test_indices:
    try:
        # Neue Berechnung mit Morley-Walter
        P_theory_mw, P_real, S_mw, Q_mw, walter_valid, kepler_limit = \
            prime_state_equation_mw(k, t_example, primes_example, 
                                   beta=0.01, window_m=25, xi=1.0, idx_sigma=10.0)
        
        # Alte Berechnung ohne Morley-Walter (zum Vergleich)
        P_theory_old, P_real_old, S_old, Q_old, _, _ = \
            prime_state_equation(k, t_example, primes_example, beta=0.01)
        
        gap = primes_example[k+1] - primes_example[k]
        ratio_old = P_theory_old / P_real if P_real > 0 else float('inf')
        ratio_mw = P_theory_mw / P_real if P_real > 0 else float('inf')
        
        print(f"\nk={k}: p[{k}]={primes_example[k]}, p[{k+1}]={primes_example[k+1]} (Gap={gap})")
        print(f"  Entropie (alt): S = {S_old:.6f}")
        print(f"  Entropie (MW):  S = {S_mw:.6f}")
        print(f"  Klassen-Resonanz Q = {Q_mw:.3f}")
        print(f"  Kepler-Limit = {kepler_limit:.6f}")
        print(f"  Walter-Restriktion: {'✓ Ja' if walter_valid else '✗ Nein'}")
        
        print(f"\n  OHNE Morley-Walter:")
        print(f"    P_theory_old = {P_theory_old:.6f}")
        print(f"    Verhältnis = {ratio_old:.6f}")
        
        print(f"\n  MIT Morley-Walter 3-Kohärenz:")
        print(f"    P_theory_mw = {P_theory_mw:.6f}")
        print(f"    Verhältnis = {ratio_mw:.6f}")
        print(f"    P_real = {P_real:.6f}")
        
        # Berechne Verbesserung
        improvement = abs(ratio_old - 1.0) - abs(ratio_mw - 1.0)
        if improvement > 0:
            print(f"\n  ✓ Verbesserung: {improvement:.6f} (näher an 1.0)")
        else:
            print(f"\n  → Weitere Optimierung möglich")
            
    except (ValueError, IndexError) as e:
        print(f"\nk={k}: Fehler - {e}")

print("\n" + "=" * 70)
```

```sage
from sage.all import *
import numpy as np

# --- 1. Die Testfunktionen (Deine Definitionen) ---
def phi_a(t, a):
    # Fourier-Transformierte der Dreiecksfunktion
    # Achtung: Numerische Stabilität bei t=0 beachten
    t_float = float(t)
    a_float = float(a)
    if abs(t_float) < 1e-9:
        return float(2.0 / a_float * a_float**2) # L'Hospital -> 2a
    term = float(sin(a_float*t_float)/t_float)
    return float((2.0/a_float) * term**2)

def F_a(x, a):
    # Kompakte Trägerfunktion
    x_float = float(x)
    a_float = float(a)
    val = float(1 - abs(x_float)/(2*a_float))
    return float(max(0, val))

def Psi_term(t):
    # Der Gamma-Integral-Term (Psi = digamma)
    # Re(psi(1/4 + it/2)) - log(pi)
    t_float = float(t)
    s_val = 0.25 + 0.5*I*t_float
    return float(-log(pi) + psi(s_val).real())

# --- 2. Der Bamberger Operator D_B ---
def build_coupled_eigenvalues(t_block, p_block, epsilon=0.01):
    N = len(t_block)
    # Konvertiere t_block zu Python-Floats für NumPy-Kompatibilität
    t_block_float = [float(t) for t in t_block]
    # Konvertiere epsilon zu Python-Float
    epsilon_float = float(epsilon)
    
    # Diagonalmatrix der originalen Nullstellen
    A = np.diag(t_block_float).astype(np.float64)
    
    # Quaternionische Kopplungsmatrix R
    R = np.zeros((N, N), dtype=np.float64)
    
    # Hilfsfunktion für Klassen
    def get_class(p):
        m = int(p) % 12
        return {5:'A', 7:'B', 11:'C', 1:'D'}.get(m, '?')
    
    classes = [get_class(p) for p in p_block]
    
    # Wir schützen die Mitte (Anker), koppeln den Rest
    mid = N // 2
    target_class = classes[mid]
    
    for i in range(N):
        for j in range(i+1, N):
            if i == mid or j == mid: continue # Topologischer Schutz
            
            # Quaternionen-Logik (vereinfacht für Test)
            c1, c2 = classes[i], classes[j]
            res = False
            # B*C=A, A*C=B, A*B=C
            if {c1, c2} == {'B', 'C'} and target_class == 'A': res = True
            if {c1, c2} == {'A', 'C'} and target_class == 'B': res = True
            if {c1, c2} == {'A', 'B'} and target_class == 'C': res = True
            
            if res:
                R[i, j] = 1.0
                R[j, i] = -1.0 # Schiefsymmetrie
                
    # Gesamtsystem A_eps = A + eps*R
    # Die Eigenwerte von D sind +/- Singulärwerte von A_eps
    # Da R schiefsymmetrisch und A diagonal reell, ist D selbstadjungiert.
    # Wir nehmen hier die Singulärwerte als die "neuen t_n"
    A_eps = A + epsilon_float * R
    # Stelle sicher, dass A_eps float64 ist
    A_eps = A_eps.astype(np.float64)
    vals = np.linalg.svd(A_eps, compute_uv=False)
    return np.sort(vals) # Sortiert für Summation

# --- 3. Der Numerische Trace-Test ---
def run_trace_test(t_list, p_list, a=2.0, epsilon=0.01):
    # Konvertiere a zu Python-Float für Kompatibilität
    a_float = float(a)
    
    # A. Berechne Primzahl-Seite (RHS) - "Die Wahrheit"
    # Summe über p, m. Da Support beschränkt auf 2a (z.B. 4),
    # reicht m*log(p) <= 2a.
    # WICHTIG: Wir beschränken die Summe auf die Primzahlen im verwendeten Block
    # für konsistente Vergleichbarkeit mit der LHS
    rhs_sum = 0.0
    
    # Summiere nur über die Primzahlen im p_list (lokal konsistent)
    for p in p_list:
        log_p = float(log(p))
        m = 1
        while True:
            arg = m * log_p
            if arg > 2*a_float: break
            
            # Term: log(p)/p^(m/2) * (F(arg) + F(-arg))
            # F ist symmetrisch -> 2 * F(arg)
            term = (log_p / (float(p)**(m/2.0))) * 2 * float(F_a(arg, a_float))
            rhs_sum += term
            m += 1
            
    # Integral Term (Numerische Integration)
    # Von -infinity bis infinity. Da phi_a schnell fällt (1/t^2), reicht -100 bis 100
    from scipy.integrate import quad
    def integrand(t):
        return float(phi_a(t, a_float) * Psi_term(t))
    
    integral_val, _ = quad(integrand, -1000, 1000)
    rhs_integral = float((1.0/(2*pi)) * integral_val)
    
    # Randterme Phi(1) + Phi(0)
    # phi_a(-i/2) und phi_a(i/2). Da phi symmetrisch: 2 * phi_a(i/2)
    # Wir nutzen die explizite Definition, Achtung mit komplexen Argumenten
    # phi_a(z) = 2/a * (sin(az)/z)^2. 
    # sin(ia/2) = i sinh(a/2). z = i/2.
    # term = (i sinh(a/2) / (i/2))^2 = (2 sinh(a/2))^2 = 4 sinh^2(a/2)
    # phi = 2/a * 4 sinh^2(a/2)
    val_boundary = float((2.0/a_float) * (2 * sinh(a_float/2))**2) # Phi(0)
    # Phi(1) ist symmetrisch dazu in der Weil-Form für F_a
    # Hier vereinfacht: Wir nehmen an User hat die Terme korrekt.
    # Wir nutzen Standard: Phi(0)+Phi(1)
    rhs_total = float(val_boundary + val_boundary - rhs_sum - rhs_integral)

    # B. Berechne Spektral-Seite (LHS) mit BAMBERGER KOPPLUNG
    # Wir nehmen die t_list, koppeln sie und summieren phi_a(t_neu)
    # Um Randeffekte zu minimieren, sollte t_list groß sein
    t_coupled = build_coupled_eigenvalues(t_list, p_list, epsilon)
    
    lhs_sum = 0.0
    for t_val in t_coupled:
        # Summe über rho und rho_bar -> 2 * phi_a(t) da phi gerade
        lhs_sum += float(2 * phi_a(float(t_val), a_float))
    
    num_coupled = len(t_coupled)
    residual = float(lhs_sum - rhs_total)
    
    # Zusätzliche Diagnose-Informationen für bessere Interpretation
    return residual, lhs_sum, rhs_total, rhs_sum, rhs_integral, val_boundary, num_coupled

# --- Testlauf ---
# Mock-Daten (In echt: zeros1.gz laden)
# Wir simulieren einen Block um 1000
p_mock = list(primes(1000, 2000))

# VERBESSERTE Näherung für t-Werte
# Die k-te Nullstelle t_k kann genähert werden durch:
# t_k ≈ (k * log(k)) / (2π) + O(1)
# wobei k der Index der Nullstelle ist (nicht der Primzahl!)

def approximate_zeta_zero_index(p):
    """Schätzt den Index k der Nullstelle, die zu Primzahl p gehört"""
    # Die k-te Primzahl ist etwa p_k ≈ k * log(k)
    # Umgekehrt: k_primzahl ≈ p / log(p) für große p
    # Aber wir brauchen den Index der Nullstelle, nicht der Primzahl
    # 
    # Beziehung zwischen Primzahl-Index und Nullstellen-Index:
    # Die k-te Nullstelle t_k ≈ (k * log(k)) / (2π)
    # Die k-te Primzahl p_k ≈ k * log(k)
    # Also: t_k ≈ p_k / (2π) für ähnliche k
    # 
    # Eine bessere Näherung: k_nullstelle ≈ p / (log(p) - log(log(p)))
    if p < 2:
        return 1
    
    # Verbesserte Näherung mit mehreren Iterationen
    log_p = float(log(p))
    # Startnäherung
    k_approx = float(p / (log_p - log(log_p)) if log_p > 1 else p / log_p)
    
    # Iterative Verbesserung (1-2 Iterationen reichen)
    for _ in range(2):
        if k_approx > 1:
            # Verwende die Umkehrung: p_k ≈ k * log(k)
            # Löse iterativ: k_new ≈ p / log(k_approx)
            k_approx = float(p / log(max(k_approx, 2)))
    
    return max(1, int(k_approx))

def approximate_t_from_prime(p):
    """Berechnet eine bessere Näherung für t basierend auf der Primzahl"""
    # Finde den geschätzten Index k der zugehörigen Nullstelle
    k_est = approximate_zeta_zero_index(p)
    
    # Verbesserte Näherung: t_k ≈ (k * log(k)) / (2π) + Korrektur
    # Basierend auf der Riemann-Siegel-Formel
    log_k = float(log(max(k_est, 2)))
    t_approx = float((k_est * log_k) / (2 * pi))
    
    # Zusätzliche Korrektur basierend auf asymptotischen Entwicklungen
    # Die t_k sind typischerweise etwas größer als die einfache Näherung
    # Korrektur-Term: ~ (log(log(k)) - 1) / (2π) für große k
    if k_est > 10:
        correction_term = float((log(log_k) - 1.0) / (2 * pi))
        t_approx += correction_term
    
    # Zusätzliche Feinabstimmung basierend auf der Primzahl
    # Größere Primzahlen korrelieren mit größeren t-Werten
    # Eine kleine Korrektur: t ≈ t_approx * (1 + 0.1/log(p))
    log_p = float(log(p))
    fine_tuning = 1.0 + 0.1 / log_p
    t_approx *= fine_tuning
    
    return t_approx

# Generiere verbesserte t-Werte
print("Generiere verbesserte Mock-Daten...")
t_mock = [approximate_t_from_prime(p) for p in p_mock]

# OPTIONAL: Wenn echte Nullstellen bereits berechnet wurden, passe Mock-Daten an
# Dies ermöglicht einen fairen Vergleich
try:
    if 't_real' in globals() and len(t_real) > 0:
        # Skaliere Mock-Daten, um im ähnlichen t-Bereich zu liegen
        t_mock_min, t_mock_max = min(t_mock), max(t_mock)
        t_real_min, t_real_max = min(t_real), max(t_real)
        
        # Option 1: Skaliere linear (behält relative Abstände)
        if t_mock_max > t_mock_min:
            scale_factor = (t_real_max - t_real_min) / (t_mock_max - t_mock_min)
            offset = t_real_min - t_mock_min * scale_factor
            t_mock_scaled = [t * scale_factor + offset for t in t_mock]
            
            print(f"\n⚠ Mock-Daten wurden an echte Nullstellen angepasst:")
            print(f"  Original t-Bereich: [{t_mock_min:.2f}, {t_mock_max:.2f}]")
            print(f"  Skalierter t-Bereich: [{min(t_mock_scaled):.2f}, {max(t_mock_scaled):.2f}]")
            print(f"  Echte t-Bereich: [{t_real_min:.2f}, {t_real_max:.2f}]")
            print(f"  → Verwende skalierte Mock-Daten für fairen Vergleich")
            t_mock = t_mock_scaled
except:
    pass  # Falls t_real nicht existiert, verwende normale Mock-Daten

print(f"\nMock-Daten generiert:")
print(f"  {len(p_mock)} Primzahlen im Bereich [{min(p_mock)}, {max(p_mock)}]")
print(f"  t-Bereich: [{min(t_mock):.2f}, {max(t_mock):.2f}]")
print(f"  (Verbesserte Näherung basierend auf Riemann-Siegel-Formel)") 

res, lhs, rhs, rhs_sum, rhs_integral, val_boundary, num_coupled = run_trace_test(t_mock, p_mock, a=2.0, epsilon=0.05)

print("=" * 60)
print("TRACE-FORMEL TEST - Detaillierte Analyse")
print("=" * 60)
print(f"\nLHS (Spektrale Seite - Summe über Nullstellen):")
print(f"  LHS = Σ φ_a(t_n) = {lhs:.6f}")
print(f"  Anzahl gekoppelter Nullstellen: {num_coupled}")
print(f"\nRHS (Arithmetische Seite - Zerlegung):")
print(f"  RHS_total = {rhs:.6f}")
print(f"    ├─ Randterme (2×Φ(0)): {2*val_boundary:.6f}")
print(f"    ├─ Primzahl-Summe (-Σ): {-rhs_sum:.6f}")
print(f"    └─ Integral-Term (-I): {-rhs_integral:.6f}")
print(f"\nResiduum (LHS - RHS): {res:.6e}")
print(f"\nRelative Abweichung: {abs(res)/max(abs(lhs), abs(rhs))*100:.2f}%")
print("\n" + "=" * 60)
print("INTERPRETATION:")
print("=" * 60)
print(f"• Die LHS summiert über {num_coupled} gekoppelte Nullstellen")
print(f"  (aus {len(t_mock)} ursprünglichen Mock-Nullstellen)")
print(f"• Die RHS summiert über {len(p_mock)} Primzahlen im Block")
print(f"• Das Residuum zeigt die Abweichung zwischen:")
print(f"  - Spektraler Seite (Nullstellen) und")
print(f"  - Arithmetischer Seite (Primzahlen + Korrekturterme)")
print(f"\nMögliche Ursachen für das Residuum:")
print(f"  ✓ Die Mock-Daten t_mock sind nur Näherungen")
print(f"  ✓ Die Randterme und Integral-Terme sind global,")
print(f"    während LHS/RHS nur lokal berechnet werden")
print(f"  ✓ Die Quaternionen-Kopplung verändert die Eigenwerte")
print(f"\nFür eine exakte Trace-Formel bräuchte man:")
print(f"  → Echte Riemann-Nullstellen (aus zeros1.gz)")
print(f"  → Alle Nullstellen bis zu einem T (nicht nur Block)")
print(f"  → Korrekte globale Randterme für den gewählten T")
```
