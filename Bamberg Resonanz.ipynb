# Bamberg Resonanz

Dieses Notebook ist die Notebook-Version von `Bamberg Resonanz.py`.

Ziel: die "Bamberger" Gitter-Resonanz mit SageMath zu testen, indem die partielle Summe
\[
\sum_{n=1}^{n_{\max}} \frac{1}{n^2}
\]
mit dem Idealwert
\[
\zeta(2)=\frac{\pi^2}{6}
\]
verglichen wird.

Ausgelegt fuer SageMath 10.8.

```python
# SageMath 10.8: Pruefung der "Bamberger" Gitter-Resonanz
from sage.all import *

def projektion_bamberg(n_max):
    # Summe der Kehrwerte der Quadrate bis n_max (lokale Dichte)
    lokale_dichte = sum(1 / n^2 for n in range(1, n_max + 1))
    fehler = abs(zeta(2) - lokale_dichte)
    return lokale_dichte.n(), fehler.n()

# Beispiel fuer n = 7
n_max = 7
dichte, diff = projektion_bamberg(n_max)
print(f"Dichte bei Iteration {n_max}: {dichte}")
print(f"Abweichung zum idealen pi^2/6: {diff}")
```

```python
# Interaktive Variante fuer SageMath 10.8
@interact
def _(n_max=slider(1, 100, 1, default=7, label='n_max')):
    dichte, diff = projektion_bamberg(n_max)
    print(f"Dichte bei Iteration {n_max}: {dichte}")
    print(f"Abweichung zum idealen pi^2/6: {diff}")
```
