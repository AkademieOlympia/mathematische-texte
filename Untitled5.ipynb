```python
# SageMath 10.5 Code zur Berechnung der Summe (a^2 - k^2)
def berechne_summe_a(a_wert):
    k = var('k')
    a = var('a')
    
    # Definition der Summenformel
    summe_formel = sum(a^2 - k^2, k, 1, a)
    
    # Vereinfachung des Ausdrucks
    vereinfacht = summe_formel.full_simplify()
    
    # Ergebnis für den spezifischen Wert
    ergebnis_wert = vereinfacht.subs(a=a_wert)
    
    return vereinfacht, ergebnis_wert

# Beispiel für eine ungerade Zahl, z.B. a = 7
ungerade_zahl = 77777
allgemeine_form, wert = berechne_summe_a(ungerade_zahl)

print(f"Allgemeine Formel für S(a): {allgemeine_form}")
print(f"Ergebnis für a = {ungerade_zahl}: {wert}")
```
